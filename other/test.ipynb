```python
import requests
import json
from bs4 import BeautifulSoup

SBA_SEARCH_URL = "https://web.sba.gov/pro-net/search/dsp_profile.cfm";
```

```python
#call sba.gov search api
SBA_response = requests.get(SBA_SEARCH_URL, params={'duns': "800751112"}).text
soup = BeautifulSoup(SBA_response, "html.parser")
profile_divs = soup.find_all("div", {"class":"profileline"})
print(profile_divs)
```

```python
for profile_div in profile_divs:
    print(profile_div.find("div", {"class":"profilehead"}).text)
    print(profile_div.find("div", {"class":"profileinfo"}).text)
```
